```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
file_path = 'https://raw.githubusercontent.com/shreyasjha5706/dumbhomework/main/data_science/JaipurFinalCleanData.csv'

df = pd.read_csv(file_path)
```

```python
display(df.head(20))
```

```python
print(df.dtypes)
```

```python
plt.scatter(df['date'], df['mean_temperature'])
plt.title('Mean Temperature Changes Over the Years in Jaipur')
plt.xlabel('Year')
plt.ylabel('Mean Temperature')
plt.show()
```

```python
plt.plot(df['date'], df['min_temperature'], label='Min Temperature')
plt.plot(df['date'], df['max_temperature'], label='Max Temperature')
plt.title('Minimum and Maximum Temperature Over the Years in Jaipur')
plt.xlabel('Year')
plt.ylabel('Temperature')
plt.legend()
plt.show()
```

```python
# Convert 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])

subset_df = df[(df['date'] >= '2016-01-01') & (df['date'] <= '2018-01-01')]
plt.hist(subset_df['mean_temperature'], bins=20, edgecolor='black')
plt.title('Temperature Distribution (2016 - 2018) in Jaipur')
plt.xlabel('Mean Temperature')
plt.ylabel('Frequency')
plt.show()
```
